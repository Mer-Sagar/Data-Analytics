# IQR = Q3-Q1

```python
import math

# x=[40,333,12,1,-500,34,67,845,65,-340]
x = [27,25,20,15,30,35,28,25,100]
l=len(x)
x.sort()
print(x)

index_Q1 = (25/100)*l
index_Q3 = (75/100)*l

if(index_Q1%1 == 0):

    Q1 = ( x[index_Q1 - 1] + x[index_Q1] )/2
    
    print("For Normal Integer Q1 is :--> ",Q1)    

else:                                                # point vali value mate
    index_Q1 = math.ceil(index_Q1)
    Q1 = x[index_Q1 - 1]
    
    print("For Float Values Q1 is :--> ",Q1)
    
    
if(index_Q3%1 == 0):

    Q3 = ( x[index_Q3 - 1] + x[index_Q3] )/2
    
    print("For Normal Integer Q3 is :--> ",Q3)
    
    
else:                                                # point vali value mate
    index_Q3 = math.ceil(index_Q3)
    Q3 = x[index_Q3 - 1]
    
    print("For Float Values Q3 is :--> ",Q3)    
    

# print(Q1,Q3)

iqr = Q3-Q1
print("IQR is :--> ",iqr)

upper = Q3 + (1.5*iqr)
lower = Q1 - (1.5*iqr)

outlier = []
lst=[]

for i in range(l):
    
    if (x[i]>upper or x[i]< lower):
        outlier.append(x[i])
    else:
        lst.append(x[i])
        
print("Outliers :--> ",outlier)
print("Data without Outlier :--> ", lst)

```

```python
import matplotlib.pyplot as plt

plt.boxplot(lst,vert=False)
```

# Variance & Z score

```python
import pandas as pd

df = pd.read_csv("Book1.csv")

m = sum(df['x'])/ df.shape[0]
devi = []
squ_devi = []


for i in range(df.shape[0]):
    
    devi.append(df.x[i] - m)
    squ_devi.append((df.x[i] - m)**2)   
    


df["Xi - Mu"] = devi
df["(Xi - Mu)**2"] = squ_devi


variance = sum(squ_devi) / df.shape[0]

std_devi = variance**(1/2)


z=[]

for i in range(df.shape[0]):
      
    z.append( devi[i] / std_devi  )
    
df["Z Score"] = z    

print("Variance is :-->",variance)
print("Standard Deviation is :-->", std_devi)
print("Cofficent of Variance is :-->", (variance**(1/2) / m)*100 )

df
```

# Covariance

```python
import pandas as pd

df2 = pd.read_csv("Covariance.csv")

x_mean = sum(df2['x'])/df2.shape[0]
y_mean = sum(df2['y'])/df2.shape[0]

x_mean
y_mean

x1,y1 = [],[]
x1_y1=[]

xs,ys=[],[]

for i in range(df2.shape[0]):
    
    x1.append( df2.x[i] - x_mean )
    y1.append( df2.y[i] - y_mean )
    x1_y1.append(x1[i]*y1[i])
    
    xs.append( x1[i]**2 )
    ys.append( y1[i]**2 )
    
    
    

df2["xi - x_mean"] = x1
df2["(Xi - X_mean )** 2"] = xs
df2["Yi - Y_mean"] = y1
df2["(Yi - Y_mean )** 2"] = ys
df2["(Xi - X_mean)(Yi - Y_mean)"] = x1_y1

n = df2.shape[0]

Sx = (sum(xs)/n)**1/2
Sy = (sum(ys)/n)**1/2

covariance =  sum(x1_y1)/n
corr_coff = covariance / (Sx * Sy)


print("Covariance is :-->",covariance)
print("Correlation Coefficient is :--> ", corr_coff)
df2
```
